```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
file_path = "Extended_Soil_Data.csv"
df = pd.read_csv(file_path)

# *Data Cleaning: Handle missing values*
df = df.dropna()

# Selecting relevant features
X = df[['moisture', 'temperature', 'r', 'g', 'b']]
y_soil = df['Soil_Type_Encoding']
y_crop1 = df['Crop_Type_Encoding']
y_crop2 = df['Crop_Type_Encoding.1']
y_crop3 = df['Crop_Type_Encoding.2']

# Splitting the dataset
X_train, X_test, y_soil_train, y_soil_test = train_test_split(X, y_soil, test_size=0.18, random_state=42)
X_train, X_test, y_crop1_train, y_crop1_test = train_test_split(X, y_crop1, test_size=0.18, random_state=42)
X_train, X_test, y_crop2_train, y_crop2_test = train_test_split(X, y_crop2, test_size=0.18, random_state=42)
X_train, X_test, y_crop3_train, y_crop3_test = train_test_split(X, y_crop3, test_size=0.18, random_state=42)

# Training Random Forest models
params = {
    'n_estimators': 100,
    'random_state': 42
}

rf_soil = RandomForestClassifier(**params)
rf_crop1 = RandomForestClassifier(**params)
rf_crop2 = RandomForestClassifier(**params)
rf_crop3 = RandomForestClassifier(**params)

rf_soil.fit(X_train, y_soil_train)
rf_crop1.fit(X_train, y_crop1_train)
rf_crop2.fit(X_train, y_crop2_train)
rf_crop3.fit(X_train, y_crop3_train)

# *Create mapping dictionaries to convert encodings back to actual values*
soil_mapping = dict(zip(df['Soil_Type_Encoding'], df['soil type']))
crop_mapping = dict(zip(df['Crop_Type_Encoding'], df['crop 1']))
crop_mapping2 = dict(zip(df['Crop_Type_Encoding.1'], df['crop 2']))
crop_mapping3 = dict(zip(df['Crop_Type_Encoding.2'], df['crop 3']))

# *User Input for Predictions*
print("Enter the values for prediction:")
moisture = float(input("Moisture: "))
temperature = float(input("Temperature: "))
r = float(input("R Value: "))
g = float(input("G Value: "))
b = float(input("B Value: "))

# Convert input into DataFrame for prediction
input_data = pd.DataFrame([[moisture, temperature, r, g, b]], columns=['moisture', 'temperature', 'r', 'g', 'b'])

# Predict using trained models
soil_prediction = rf_soil.predict(input_data)[0]
crop1_prediction = rf_crop1.predict(input_data)[0]
crop2_prediction = rf_crop2.predict(input_data)[0]
crop3_prediction = rf_crop3.predict(input_data)[0]

# Convert numerical predictions back to actual names
soil_actual = soil_mapping.get(soil_prediction, "Unknown")
crop1_actual = crop_mapping.get(crop1_prediction, "Unknown")
crop2_actual = crop_mapping2.get(crop2_prediction, "Unknown")
crop3_actual = crop_mapping3.get(crop3_prediction, "Unknown")

# Print actual predictions
print(f"Predicted Soil Type: {soil_actual}")
print(f"Predicted Crop 1: {crop1_actual}")
print(f"Predicted Crop 2: {crop2_actual}")
print(f"Predicted Crop 3: {crop3_actual}")

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
file_path = "Extended_Soil_Data.csv"
df = pd.read_csv(file_path)

# *Data Cleaning: Handle missing values*
df = df.dropna()

# Selecting relevant features
X = df[['moisture', 'temperature', 'r', 'g', 'b']]
y_soil = df['Soil_Type_Encoding']
y_crop1 = df['Crop_Type_Encoding']
y_crop2 = df['Crop_Type_Encoding.1']
y_crop3 = df['Crop_Type_Encoding.2']

# Splitting the dataset
X_train, X_test, y_soil_train, y_soil_test = train_test_split(X, y_soil, test_size=0.18, random_state=42)
X_train, X_test, y_crop1_train, y_crop1_test = train_test_split(X, y_crop1, test_size=0.18, random_state=42)
X_train, X_test, y_crop2_train, y_crop2_test = train_test_split(X, y_crop2, test_size=0.18, random_state=42)
X_train, X_test, y_crop3_train, y_crop3_test = train_test_split(X, y_crop3, test_size=0.18, random_state=42)

# Training Random Forest models
params = {
    'n_estimators': 100,
    'random_state': 42
}

rf_soil = RandomForestClassifier(**params)
rf_crop1 = RandomForestClassifier(**params)
rf_crop2 = RandomForestClassifier(**params)
rf_crop3 = RandomForestClassifier(**params)

rf_soil.fit(X_train, y_soil_train)
rf_crop1.fit(X_train, y_crop1_train)
rf_crop2.fit(X_train, y_crop2_train)
rf_crop3.fit(X_train, y_crop3_train)

# *Create mapping dictionaries to convert encodings back to actual values*
soil_mapping = dict(zip(df['Soil_Type_Encoding'], df['soil type']))
crop_mapping = dict(zip(df['Crop_Type_Encoding'], df['crop 1']))
crop_mapping2 = dict(zip(df['Crop_Type_Encoding.1'], df['crop 2']))
crop_mapping3 = dict(zip(df['Crop_Type_Encoding.2'], df['crop 3']))

# *User Input for Predictions*
print("Enter the values for prediction:")
moisture = float(input("Moisture: "))
temperature = float(input("Temperature: "))
r = float(input("R Value: "))
g = float(input("G Value: "))
b = float(input("B Value: "))

# Convert input into DataFrame for prediction
input_data = pd.DataFrame([[moisture, temperature, r, g, b]], columns=['moisture', 'temperature', 'r', 'g', 'b'])

# Predict using trained models
soil_prediction = rf_soil.predict(input_data)[0]
crop1_prediction = rf_crop1.predict(input_data)[0]
crop2_prediction = rf_crop2.predict(input_data)[0]
crop3_prediction = rf_crop3.predict(input_data)[0]

# Convert numerical predictions back to actual names
soil_actual = soil_mapping.get(soil_prediction, "Unknown")
crop1_actual = crop_mapping.get(crop1_prediction, "Unknown")
crop2_actual = crop_mapping2.get(crop2_prediction, "Unknown")
crop3_actual = crop_mapping3.get(crop3_prediction, "Unknown")

# Print actual predictions
print(f"Predicted Soil Type: {soil_actual}")
print(f"Predicted Crop 1: {crop1_actual}")
print(f"Predicted Crop 2: {crop2_actual}")
print(f"Predicted Crop 3: {crop3_actual}")

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
file_path = "Extended_Soil_Data.csv"
df = pd.read_csv(file_path)

# *Data Cleaning: Handle missing values*
df = df.dropna()

# Selecting relevant features
X = df[['moisture', 'temperature', 'r', 'g', 'b']]
y_soil = df['Soil_Type_Encoding']
y_crop1 = df['Crop_Type_Encoding']
y_crop2 = df['Crop_Type_Encoding.1']
y_crop3 = df['Crop_Type_Encoding.2']

# Splitting the dataset
X_train, X_test, y_soil_train, y_soil_test = train_test_split(X, y_soil, test_size=0.18, random_state=42)
X_train, X_test, y_crop1_train, y_crop1_test = train_test_split(X, y_crop1, test_size=0.18, random_state=42)
X_train, X_test, y_crop2_train, y_crop2_test = train_test_split(X, y_crop2, test_size=0.18, random_state=42)
X_train, X_test, y_crop3_train, y_crop3_test = train_test_split(X, y_crop3, test_size=0.18, random_state=42)

# Training Random Forest models
params = {
    'n_estimators': 100,
    'random_state': 42
}

rf_soil = RandomForestClassifier(**params)
rf_crop1 = RandomForestClassifier(**params)
rf_crop2 = RandomForestClassifier(**params)
rf_crop3 = RandomForestClassifier(**params)

rf_soil.fit(X_train, y_soil_train)
rf_crop1.fit(X_train, y_crop1_train)
rf_crop2.fit(X_train, y_crop2_train)
rf_crop3.fit(X_train, y_crop3_train)

# *Create mapping dictionaries to convert encodings back to actual values*
soil_mapping = dict(zip(df['Soil_Type_Encoding'], df['soil type']))
crop_mapping = dict(zip(df['Crop_Type_Encoding'], df['crop 1']))
crop_mapping2 = dict(zip(df['Crop_Type_Encoding.1'], df['crop 2']))
crop_mapping3 = dict(zip(df['Crop_Type_Encoding.2'], df['crop 3']))

# *User Input for Predictions*
print("Enter the values for prediction:")
moisture = float(input("Moisture: "))
temperature = float(input("Temperature: "))
r = float(input("R Value: "))
g = float(input("G Value: "))
b = float(input("B Value: "))

# Convert input into DataFrame for prediction
input_data = pd.DataFrame([[moisture, temperature, r, g, b]], columns=['moisture', 'temperature', 'r', 'g', 'b'])

# Predict using trained models
soil_prediction = rf_soil.predict(input_data)[0]
crop1_prediction = rf_crop1.predict(input_data)[0]
crop2_prediction = rf_crop2.predict(input_data)[0]
crop3_prediction = rf_crop3.predict(input_data)[0]

# Convert numerical predictions back to actual names
soil_actual = soil_mapping.get(soil_prediction, "Unknown")
crop1_actual = crop_mapping.get(crop1_prediction, "Unknown")
crop2_actual = crop_mapping2.get(crop2_prediction, "Unknown")
crop3_actual = crop_mapping3.get(crop3_prediction, "Unknown")

# Print actual predictions
print(f"Predicted Soil Type: {soil_actual}")
print(f"Predicted Crop 1: {crop1_actual}")
print(f"Predicted Crop 2: {crop2_actual}")
print(f"Predicted Crop 3: {crop3_actual}")
```
